**Select on 1951-2003:**

```python
data = {}
for varname,_varname_fred in varspecs:
    years = data_fred[f'{varname}'].index.year
    I = (years >= 1951) & (years <= 2003)
    data[varname] = data_fred[f'{varname}'][I].values
```

**Construct quarterly wage rate data**

```python
data_temp1 = pd.DataFrame(data_fred['hours'])
data_temp1.columns = ['hours']
data_temp1['Year'] = data_temp1.index.year

data_temp2 = pd.DataFrame(data_fred['wage_tot'])
data_temp2.columns = ['wage_tot']
data_temp2['Year'] = data_temp2.index.year

data_merged = pd.merge(data_temp1.assign(grouper=data_temp1['Year']),
             data_temp2.assign(grouper=data_temp2['Year']),
             how='left', on='grouper', left_index= True)

data_merged['wage_rate'] = data_merged['wage_tot']/(data_merged['hours']/4)

# b. select on year
years = data_merged.index.year     
I = (years >= 1951) & (years <= 2003)
data['w'] = data_merged[f'wage_rate'][I].values
```

# Convert to quarterly frequency

```python
for varname in ['u','pop','vt','CPI']:
    convert_m_to_qtr(data,varname)
```

**Calculate real quarterly and wage rate and GDP per capita**:

```python
data['w_qtr'] = data['w']/(data['deflator'])
data['y_qtr'] = data['GDP']/(data['deflator']*data['pop_qtr'])
data['a_qtr'] = data['productivity_nonfarm']
```

**Calculate nominal quarterly interest rate:**

```python
R_qtr = data['R_qtr']
data['R_qtr'] = np.nan*np.ones(data['a_qtr'].size)
data['R_qtr'][14:] = R_qtr
```

**Calculate annualized quarterly inflaiton:**

```python
data['Pi_qtr'] = np.nan*np.ones(data['a_qtr'].size)
data['Pi_qtr'][1:] = (data['deflator'][1:]/data['deflator'][:-1])**4-1
data['Pi_CPI_qtr'] = np.nan*np.ones(data['a_qtr'].size)
data['Pi_CPI_qtr'][1:] = (data['CPI_qtr'][1:]/data['CPI_qtr'][:-1])**4-1
```

# Calculate job finding and job seperation rates

```python
# a. adjustment
years = data_fred[f'unemp'].index.year
I = (years >= 1951) & (years <= 2003)
J = years[I] >= 1993
data['unemp_short'][J] *= 1.1 # adjustment suggested by Shimer (2005)
data['unemp'][J] *= 1.1 # adjustment suggested by Shimer (2005)

# b. formulas
data['lambda_u'] = np.nan*np.ones(data['unemp'].size)
data['deltat'] = np.nan*np.ones(data['unemp'].size)
data['lambda_u'][:-1] = 1.0-(data['unemp'][1:]-data['unemp_short'][1:])/data['unemp'][:-1]
data['deltat'][:-1] = data['unemp_short'][1:]/(data['emp'][:-1]*(1-0.5*data['lambda_u'][:-1]))
```

Convert to quarterly:

```python
for varname in ['lambda_u','deltat']:
    convert_m_to_qtr(data,varname)
```

# Apply HP filter and calculate moments

```python
datamoms = moms(data,varnames)
```

```python
with open('data/datamoms.pickle','wb') as handle:
    pickle.dump(datamoms,handle)  
```

# Show moments and plots

```python
qtrs = np.arange('1951-01', '2003-12', dtype='datetime64[M]')[::3]
```

```python
for varname in varnames:
    
    # a. moments
    mean = datamoms[('mean',f'{varname}')]
    std = datamoms[('std',f'{varname}')]
    autocorr1 = datamoms[('autocorr1',f'{varname}')]
    corr0_u = datamoms[('corr0','u_qtr',f'{varname}')]
    corr0_a = datamoms[('corr0','a_qtr',f'{varname}')]
    
    print(f'{varname}: mean = {mean:.3f}, std. = {std:.3f}, autocorr = {autocorr1:.3f}, corr(u,x) = {corr0_u:.3f}, corr(a,x) = {corr0_a:.3f}')
    
    # b. plot
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(qtrs,data[varname],label='raw data')
    ax.plot(qtrs,np.exp(data[f'{varname}_hp_trend']),label='HP trend')
    ax.legend(frameon=True)
    plt.show()
```
